# Run

```python
category =  "config2"
run = 46
name = "S"
fsamp = .02
```

```python
data = pd.read_csv(f"{home_dir}/CHARLES/{category}/R{run}/{name}.csv")
data = data.iloc[100:]
N = len(data)
plt.figure()
plt.plot(np.linspace(0, N*fsamp, N), data)
```

```python

f, E = sp.signal.welch(data, fs = fsamp, axis = 0, nperseg = N//4, scaling = 'density', detrend = 'constant') 

plt.figure()
plt.loglog(f, E)
plt.xlabel("frequency $[1/s]$")
```
